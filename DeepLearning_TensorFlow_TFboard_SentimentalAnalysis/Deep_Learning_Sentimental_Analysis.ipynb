# install by using pip install tflearn

```python
import tflearn
```

```python
from tflearn.data_utils import  to_categorical,pad_sequences
from tflearn.datasets import imdb
```

```python
#import the data from imdb into the local dataset
```

```python
train,test,_=imdb.load_data(path='imdb.pkl',n_words=10000,valid_portion=0.1)
#n_words= no of words that needs to be imported from the database
#valid_portion=0.1 means total data imported into 10% validation 90%training
#.pkl is file which is useful fo converting into lists easily
```

```python
trainX,trainY=train
testX,testY=test
```

# Vectorize the inputs as you cant pass the text into the NN directly NN 



```python
trainX=pad_sequences(trainX,maxlen=100,value=0.)
testX=pad_sequences(testX,maxlen=100,value=0.)
```

# Now to convert the output you need to check the other things...As our output is categerical which means postive or negative
#so we convert our Y values into the categorical values


```python
trainY=to_categorical(trainY,nb_classes=2)
testY=to_categorical(testY,nb_classes=2)
```

# NN Start here

```python
net=tflearn.input_data([None,100])
#As we given the maxlen=100 in trainX which is our input
```

# Output of Previous layer as a input to the next layer
input dimension=10000 =no fof words you loaded from the dataset
no of dimesions of resulting embedding=128

```python

net=tflearn.embedding(net,input_dim=10000,output_dim=128)
```

# Next feed that layer to the lstm so it will remmeber and improve prediction
Dropout is to prevent over fitting a.Randomly dropping the NN and forcing the 
neurons to learn from the other neurons

```python
net=tflearn.lstm(net,128,dropout=0.8)
```

# Fully Connected layer which means  outputs of each previpus layer is connceted to  all the layers of the next layer


```python
net=tflearn.fully_connected(net,2,activation='softmax')
```

# adam=which performs the gradient descent
##categorical_crossentropy=is the loss between predicted and actual output

```python
net=tflearn.regression(net,optimizer='adam',learning_rate=0.00001,loss='categorical_crossentropy')
```

# Training-by tlearn DeepNueralNet Function(DFF)
show_metric=True View the log of accuracy during the training method

```python
model=tflearn.DNN(net,tensorboard_verbose=0)
```

```python
model.fit(trainX,trainY,validation_set=(testX,testY),show_metric=True,batch_size=32)
```
